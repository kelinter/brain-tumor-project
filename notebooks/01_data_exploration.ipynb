```python
import os
from collections import Counter
import matplotlib.pyplot as plt
import cv2

base_dir = "../data/raw"    
classes = sorted(os.listdir(base_dir))
counts = {cls: len(os.listdir(f"{base_dir}/{cls}")) for cls in classes}
print("Image counts:", counts)

fig, axes = plt.subplots(1, len(classes), figsize=(12, 6))
for ax, cls in zip(axes, classes):
    img_path = os.path.join(base_dir, cls, os.listdir(f"{base_dir}/{cls}")[0])
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    ax.imshow(img, cmap="gray")
    ax.set_title(f"{cls} ({counts[cls]})")
    ax.axis("off")
plt.show()
```
